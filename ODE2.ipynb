```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# define the differential equation
def model(y, x):
    dydx = [y[1], (x**2 - 5*y[0])/y[0]]
    return dydx

# initial conditions
y0 = [1, 0]

# x values
x = np.linspace(0, 10, 100)

# solve the differential equation
sol = odeint(model, y0, x)

# plot the solution
plt.plot(x, sol[:, 0])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution to Differential Equation')
plt.grid()
plt.show()
```

## The values of x's and y's are given below

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# define the differential equation
def model(y, x):
    dydx = [y[1], (x**2 - 5*y[0])/y[0]]
    return dydx

# initial conditions
y0 = [1, 0]

# x values
x = np.linspace(0, 10, 100)

# solve the differential equation
sol = odeint(model, y0, x)

# print the values of x and y
for i in range(len(x)):
    print(f"x = {x[i]:.2f}, y = {sol[i, 0]:.4f}")

# plot the solution
plt.plot(x, sol[:, 0])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution to Differential Equation')
plt.grid()
plt.show()
```
