```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the model
def model(x, y):
    dydx = np.sqrt(x**2 - y)
    return dydx

# Set initial condition and x values
y0 = 0
x_span = (0, 10)
x = np.linspace(0, 10, 100)

# Solve ODE
sol = solve_ivp(model, x_span, [y0], t_eval=x, method='RK45')

# Print x and y values
for i in range(len(x)):
    print(f"x = {x[i]}, y = {sol.y[0][i]}")

# Plot x and y values
plt.figure(figsize=(10, 4))
plt.plot(x, sol.y[0], label='y')
plt.plot(x, -sol.y[0], label='-y')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Solution of (y')^2 + y = x^2")
plt.legend()
plt.grid()
plt.show()
```
