```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(y, x):
 dydt = (x**2 - 5*y)/y
 return dydt

# initial condition
y0 = 1

# time points
x = np.linspace(0, 10)

# solve ODE
sol = odeint(model, y0, x)

# plot results
plt.plot(x, sol, 'b', label='y(x)')
plt.legend(loc='best')
plt.xlabel('x')
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(y, x):
 dydt = (x**2 - 5*y)/y
 return dydt

# initial condition
y0 = 1

# time points
x = np.linspace(0, 10, 100)

# solve ODE
sol = odeint(model, y0, x)

# print values of x and y
for i in range(len(x)):
 print(f"x = {x[i]:.4f}, y = {sol[i][0]:.4f}")

# plot results
plt.plot(x, sol, 'b', label='y(x)')
plt.legend(loc='best')
plt.xlabel('x')
plt.grid()
plt.show()
```
