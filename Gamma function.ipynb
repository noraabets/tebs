```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

# Create an array of x values
x = np.linspace(0.1, 5, 100)

# Calculate the corresponding gamma values
y = gamma(x)

# Create the plot
plt.plot(x, y)
plt.title('Gamma Function')
plt.xlabel('x')
plt.ylabel('Γ(x)')
plt.grid(True)
plt.show()
```
